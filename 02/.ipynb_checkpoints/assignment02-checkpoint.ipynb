```python
import os


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

make array and initilize value

make input and generate noise value 

```python
leng = 10 # [0,10]
datanum= leng * 100#트레인 데이터 갯수

epoch =10000


loss_Array=np.zeros((epoch,1))
W_Array=np.zeros((epoch,1))
B_Array=np.zeros((epoch,1))

x_=np.zeros((datanum,1))
y_=np.zeros((datanum,1))
noise_y=np.zeros((datanum,1))
new_y=np.zeros((datanum,1))

```

```python


origin_W = 2 
origin_B = 3


for i in range(datanum):
    x_[i]=i*leng/datanum
    y_[i]=origin_W*x_[i]+origin_B
    noise_y[i]=np.random.normal(y_[i],2)
    
    
```

training

```python
lr=0.01 #learning Rate

W=0.001
B=0.001

for iter in range(epoch):
    print("iter :" + str(iter+1))
    
    for i in range(datanum):
        new_y[i]=W*x_[i]+B
    
    loss=(new_y-noise_y)*(new_y-noise_y) 
    loss=1/(2*datanum)*np.sum(loss)#로스함수 평균해서 코스트 구함
    loss_Array[iter]=loss
    
    print(loss)
    
    W_ = 1/datanum*np.sum((new_y-noise_y)*x_)
    B_ = 1/datanum*np.sum(new_y-noise_y)

    W = W - W_ * lr 
    B = B - B_ * lr
    
    W_Array[iter] = W
    B_Array[iter] = B
 
```

output value

plotting the input data

```python
plt.scatter(x_,y_ ,c="r",s=2)#train loss 
plt.scatter(x_,noise_y ,c="b",s=2,alpha=0.2 )#test loss
plt.show()
```
